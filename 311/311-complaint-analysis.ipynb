```python
import pandas as pd
```

```python
## 311 idling complaints from 1/1/15 to 5/5/25- filtered for Manhattan
df = pd.read_csv("311-data.csv")
df
```

```python
## extract the year from the created_date column
## the format thing is because it was giving a warning that pandas was confused about the format, i asked chatgpt what to do

df["Created Date"] = pd.to_datetime(df["Created Date"], errors="coerce")

df["Year"] = df["Created Date"].dt.year
df["Year"]
```

```python
##trying to do the same with month as well, so can try to see data for the same months changes the findings. (Before, I was comparing 
# 12 months data from prev years to 4 months data from 2025.
df["Created Date"] = pd.to_datetime(df["Created Date"], errors="coerce")

df["Month"] = df["Created Date"].dt.month
df["Month"]
```

```python
## checking to see if 'Year' got added as a column in the df -- yes!
## gave it a list instead of creating another df
df.head()[["Year", "Month", "Cross Street 1", "Cross Street 2", "Street Name", "Incident Zip", "Incident Address", "Unique Key"]]
```

```python
## Adding a column called updated_date for mm/yyyy instead of created date
## combining month and year, used claude to help with code for lambda 

df["updated_date"] = df.apply(lambda x: f"{x['Month']:02d}/{str(x['Year'])}", axis=1)
df
```

```python
## adding seasonal df that includes data for only jan-apr  
## I only wanted mm-yyyy, this is why updated_date shows 01 as date for all rows because Pandas doesn't get day 
#information, it automatically defaults to the 1st day of the month

df['updated_date'] = pd.to_datetime(df['updated_date'], format='%m/%Y') ## this converts the column to date-time format
seasonal_df = df[df['updated_date'].dt.month.isin([1, 2, 3, 4])]
seasonal_df
```
