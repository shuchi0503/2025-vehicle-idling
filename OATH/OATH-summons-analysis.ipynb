```python
import pandas as pd
```

```python
## idling summons issued and filed at the OATH Hearings Division by City enforcement agencies from 1/1/15 to 5/5/25
## filtered for dates and violation location (borough)- Manhattan
df = pd.read_csv("oath-data.csv")
df
```

```python
df.info()
```

```python
df.value_counts("Respondent Last Name")
```

```python
## extract the year from the violation date column

df["Violation Date"] = pd.to_datetime(df["Violation Date"], errors="coerce")

df["Year"] = df["Violation Date"].dt.year
df["Year"]
```

```python
df["Violation Date"] = pd.to_datetime(df["Violation Date"], errors="coerce")

df["Month"] = df["Violation Date"].dt.month
df["Month"]
```

```python
#list to remove other columns 
df.head()[["Year", "Month", "Violation Location (Zip Code)", "Violation Location (Street Name)", "Charge #1: Code Description", "Ticket Number"]]
```

```python
## combining month and year 

df["updated_date"] = df.apply(lambda x: f"{x['Month']:02d}/{str(x['Year'])}", axis=1)
df
```

```python
## adding seasonal df that includes data for only jan-apr  

df['updated_date'] = pd.to_datetime(df['updated_date'], format='%m/%Y') ## this converts the column to date-time format
seasonal_df = df[df['updated_date'].dt.month.isin([1, 2, 3, 4])]
seasonal_df
```

```python
## using zip codes because street data is a bit confusing- doesn't have cross-street info (some mentioned within the col, some don't, some\
#incomplete- for eg 'on columbus avenue between w'), it doesn't follow any naming conventions, some street names are without proper 
#spaces etc. 

##low- looking close to congestion zone- 10069, 10065, 10023

low_seasonal_df = seasonal_df[["Year", "Month", "updated_date", "Violation Location (Zip Code)", "Violation Location (Street Name)", \
                               "Charge #1: Code Description", "Ticket Number"]].query("`Violation Location (Zip Code)` == 10069.0 or \
                                `Violation Location (Zip Code)` == 10065.0 or `Violation Location (Zip Code)` == 10023.0")
low_seasonal_df
```

```python
## mid zip codes- 10021, 10023
mid_seasonal_df = seasonal_df[["Year", "Month", "updated_date", "Violation Location (Zip Code)", "Violation Location (Street Name)", \
                               "Charge #1: Code Description", "Ticket Number"]].query("`Violation Location (Zip Code)` == 10023.0 or \
                                `Violation Location (Zip Code)` == 10021.0")
mid_seasonal_df
```

```python
## high zip codes- 10028, 10024
high_seasonal_df = seasonal_df[["Year", "Month", "updated_date", "Violation Location (Zip Code)", "Violation Location (Street Name)", \
                               "Charge #1: Code Description", "Ticket Number"]].query("`Violation Location (Zip Code)` == 10028.0 or \
                                `Violation Location (Zip Code)` == 10024.0")
high_seasonal_df
```

```python
## counting complaints in low, mid and high streets from jan-april in each year. 
## sort_index- prev it showed 2016 last

count_low = low_seasonal_df.groupby("Year").size()
count_mid = mid_seasonal_df.groupby("Year").size()
count_high = high_seasonal_df.groupby("Year").size()

total_count = pd.concat([count_low.rename("lower streets"),
                        count_mid.rename("middle streets"),
                        count_high.rename("higher streets")], axis = 1).sort_index()
total_count.reset_index()
```

```python
## download for datawrapper
total_count.to_csv("oath-findings.csv",index = True )
```
